Također su korištene pickle datoteke kako bi se ubrzala naknadna pokretanja koda pohranjivanjem parsirane `movies.txt` datoteke i gotovo izgrađenog stabla.

````python
from binaryTree import *
from loadingMovies import *
from OperacijeNaStablu import *

movies = load_movies_from_file("movie.txt", "movies.pickle")
print(f"Total movies loaded: {len(movies)}")

try:
    avlTree = loadTreeAsPickle("tree.pickle")
    print("Loaded tree from tree.pickle")
except:
    avlTree = BinaryTree()

    for movie in movies:
        if not movie in avlTree:
            avlTree.add(movie)

    saveTreeAsPickle(avlTree, "tree.pickle")
    print("Created new tree.pickle")
````
